```python
import numpy as np
import cv2
```

```python
# Load an color image in grayscale
img = cv2.imread('images/patrick.jpg',0)
```

```python
# display loaded image
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# display image in named window
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('images/patric_gray.jpg',img)
```

```python
# clone gray patrick image with keyboard command
img = cv2.imread('images/patric_gray.jpg',0)
cv2.imshow('image',img)
k = cv2.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('images/patric_gray_clone.jpg',img)
    cv2.destroyAllWindows()
```
